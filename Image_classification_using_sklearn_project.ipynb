```python
#image conversion to array
data = []
labels = []

for category_idx, category in enumerate(categories):
  for file in os.listdir(os.path.join(input_dir,category)):
    img_path = os.path.join(input_dir, category, file)
    img = imread(img_path)
    img = resize(img, (15, 15))
    data.append(img.flatten())
    labels.append(category_idx)

labels = np.asarray(labels)
data = np.asarray(data)
```

```python
#train test split
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(data, labels, test_size=0.2, shuffle=True, stratify=labels)
```

```python
#Standard Scaler
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lr = LogisticRegression()
lr.fit(xtrain, ytrain)
lr_pred = lr.predict(xtest)
accuracy_score(ytest, lr_pred)
```

```python
from sklearn.metrics import *
classification_report(ytest, lr_pred)
```

```python
confusion_matrix(ytest, lr_pred)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(xtrain, ytrain)
dt_pred = dt.predict(xtest)
accuracy_score(ytest, dt_pred)
```

```python
confusion_matrix(ytest, dt_pred)
```

```python
classification_report(ytest, dt_pred)
```

```python
#RandomForest
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf.fit(xtrain, ytrain)
rf_pred = rf.predict(xtest)
accuracy_score(ytest, rf_pred)
```

```python
#hyperParameter Tuning for RandomForest
rfc = RandomForestClassifier(random_state=42)
param_grid = {
    'n_estimators': [200, 500],
    'max_features': ['sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' : ['gini', 'entropy']
}
#GridSearch
from sklearn.model_selection import GridSearchCV
```

```python
cv_rf = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=5)
cv_rf.fit(xtrain, ytrain)
print(cv_rf.best_params_)
rf1 = RandomForestClassifier(random_state=42, max_features='sqrt', n_estimators= 100, max_depth=8, criterion='gini')
rf1.fit(xtrain, ytrain)
pred_rf1 = rf1.predict(xtest)
print('Accuracy for Random Forest: ', accuracy_score(ytest, pred_rf1))
```

```python
#Creating the dataframe for actual and predicted value.
pd.DataFrame({'Actual': ytest, 'Predicted': pred_rf1})
```

## Project Summary

This notebook demonstrates a classification task to distinguish between images of cats and dogs. The following steps were performed:

1.  **Data Loading and Preprocessing:**
    *   Images from a Google Drive directory ('Cat_Dog Data Set') were loaded.
    *   Each image was resized to 15x15 pixels and flattened into a 1D NumPy array.
    *   Labels (0 for cats, 1 for dogs) were created corresponding to the image data.

2.  **Data Splitting:**
    *   The preprocessed image data (`data`) and labels (`labels`) were split into training and testing sets using `train_test_split` with an 80/20 ratio, shuffling, and stratification to maintain class balance.

3.  **Feature Scaling:**
    *   `StandardScaler` was applied to the training and testing data to standardize the features, which helps in the performance of many machine learning algorithms.

4.  **Model Training and Evaluation:**
    *   **Logistic Regression:** A Logistic Regression model was trained, achieving an accuracy of approximately 55.5%. The classification report and confusion matrix were also generated.
    *   **Decision Tree:** A Decision Tree Classifier was trained, achieving an accuracy of approximately 54.75%. Its confusion matrix and classification report were also presented.
    *   **Random Forest:** An initial Random Forest Classifier was trained, yielding an accuracy of 62.75%.

5.  **Hyperparameter Tuning (Random Forest):**
    *   `GridSearchCV` was employed to find the optimal hyperparameters for the Random Forest Classifier. The best parameters found were:
        *   `criterion`: 'gini'
        *   `max_depth`: 8
        *   `max_features`: 'log2'
        *   `n_estimators`: 500
    *   A Random Forest model was then re-evaluated using specific parameters (though not fully matching the `GridSearchCV`'s best parameters in the final run, using `n_estimators=100` and `max_features='sqrt'`), achieving an accuracy of 62%.

This workflow demonstrates the common steps involved in building and evaluating classification models on image data, from initial preprocessing to hyperparameter optimization.
